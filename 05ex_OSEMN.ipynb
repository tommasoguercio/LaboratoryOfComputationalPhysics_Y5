4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
# nnnn nnnn nnnn nnnn1010\r\n
# 19*6 + 6 = 20*6
# tot = 120 bit per row
# ASCII: SPACE = 10 0000
#        DIGIT = 11 xxxx
#      PADDING = 10 10\n\r
# chr(x): x:dec -> ASCII char
import struct, time

with  open('credit_card.dat','rb') as file:
    file_content=file.read()
    word_size = 6
    
    for i in range(0, len(file_content), word_size):
        num = chr(int(file_content[i:i+word_size], base=2))
        print(num, end='')
```

5\. Load the file "user_data.json", filter the data by the "CreditCardType" field equals to "American Express". Than save the data a to CSV.

```python
import csv
import json

data = json.load(open('user_data.json'))

with open('user_data.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(data[0]) #header
    
    for i in range(len(data)):
        if data[i]['CreditCardType'] == "American Express":
            writer.writerow(data[i].values())
```
